```
#import the transformed data of train.csv
train = pd.read_csv('/content/drive/My Drive/Colab Notebooks/trainFormulated.csv')
```

```
#make one variable with just the SalePrice column
#and one with everything else
#to be used with the regressors later
sPrice = train['SalePrice']
notSP = train.drop(['SalePrice'],axis=1)
```

```
#import different ficntions and regressors
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import ElasticNet, Lasso, Ridge, ElasticNetCV, LassoCV, RidgeCV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.svm import SVR
from mlxtend.regressor import StackingCVRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```
#use the different regressors on the transformed data

#use lasso with gridsearch and fit it to the train data
lasso = Lasso()
lasso_gs = GridSearchCV(lasso, {'alpha': np.logspace(-4, -3, 5)}, 
                            cv=5, scoring="neg_mean_squared_error")

lasso_gs.fit(notSP, sPrice)

#best estimator chooses the estimator that gives the best score
lasso_gs.best_estimator_

#construct a pipeline from the given estimator
lasso_gs = make_pipeline(RobustScaler(), Lasso(alpha=0.0005, random_state=1))
```

```
#use ElasticNet with gridsearch and fit it to the train data
enet = ElasticNet()
enet_gs = GridSearchCV(enet, {'alpha': np.linspace(0.0001, 0.001, 10), 
                                  'l1_ratio':np.linspace(0.5, 1.5, 10)},
                           cv=5, scoring="neg_mean_squared_error")
enet_gs.fit(notSP, sPrice)
enet_gs.best_estimator_

enet_gs = make_pipeline(RobustScaler(), ElasticNet(alpha=0.0004, 
                                                   l1_ratio=1.4, random_state=3))
```

```
#use Ridge with gridsearch and fit it to the train data
ridge = Ridge()
ridge_gs = GridSearchCV(ridge, {'alpha': np.linspace(10, 30, 10)}, 
                            cv=5, scoring="neg_mean_squared_error")
ridge_gs.fit(notSP, sPrice)
ridge_gs.best_estimator_

ridge_gs = make_pipeline(RobustScaler(), Ridge(alpha=19))
```

```
#set parameters for GradientBoostingRegressor
gbdt_r = GradientBoostingRegressor(learning_rate=0.05, min_samples_leaf=5, 
                                   min_samples_split=10,max_depth=4, n_estimators=3000)

#set parameters for RandomForestRegressor
rf_r = RandomForestRegressor(min_samples_leaf=4, min_samples_split=8)

#make a pipeline with SVR
svr_r = make_pipeline(RobustScaler(), SVR(C=20, epsilon=0.005, gamma=0.0003))

#set parameters for XGBRegressor
xgb_r = XGBRegressor(learning_rate=0.01, max_depth=5, n_estimators=3000, 
                         n_thread=-1, n_jobs=-1, objective='reg:squarederror')

#set parameters for LGBMRegressor
lgb_r = LGBMRegressor(objective='regression',
                    learning_rate=0.01, max_depth=5, num_leaves=4, 
                    n_estimators=3000)

#set parameters for StackingCVRegressor
stack_r = StackingCVRegressor([lasso_gs, ridge_gs, enet_gs, 
                               gbdt_r, rf_r, svr_r, xgb_r, lgb_r], 
                                  meta_regressor=lgb_r,
                                  use_features_in_secondary=True)
```

```
#fit all the regressors to the transformed data
lasso_f = lasso_gs.fit(notSP, sPrice)
ridge_f = ridge_gs.fit(notSP, sPrice)
enet_f = enet_gs.fit(notSP, sPrice)
gbdt_f = gbdt_r.fit(notSP, sPrice)
rf_f = rf_r.fit(notSP, sPrice)
svr_f = svr_r.fit(notSP, sPrice)
xgb_f = xgb_r.fit(notSP, sPrice)
lgb_f= lgb_r.fit(notSP, sPrice)
stack_f = stack_r.fit(np.array(notSP), np.array(sPrice))
```

```
#make a prediction baes in the different models
#here try to change the precentage of the different regressors
prediction = ((0.1 * enet_f.predict(test)) + \
            (0.1 * ridge_f.predict(test)) + \
            (0.1 * lasso_f.predict(test)) + \
            (0.15 * gbdt_f.predict(test)) + \
            (0.15 * xgb_f.predict(test)) + \
            (0.1 * lgb_f.predict(test)) + \
            (0.075 * rf_f.predict(test)) + \
            (0.075 * svr_f.predict(test)) + \
            (0.15 * stack_f.predict(np.array(test)))
           )
```

```
#convert prediction into a dataframe, import the sample file
#concatinate the prediction with the Id from sample submission
#and export the data as a csv file
prediction_df = pd.DataFrame(prediction)
sample_df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/sample_submission.csv')
final_data = pd.concat([sample_df['Id'],prediction_df],axis=1)
final_data.columns = ['Id', 'Saleprice']
final_data.to_csv('final_submission.csv',index=False)
final_data.head()
```
